```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
df = pd.read_csv('../data/secom.csv')
display(df.head())
```

### Plot 1: Target Distribution
Understanding the balance of the target variable (Yield Pass/Fail).

```python
plt.figure(figsize=(6, 4))
sns.countplot(x=df.iloc[:, -1])
plt.title('Target Variable Distribution')
plt.xlabel('Target (Yield)')
plt.ylabel('Count')
plt.show()
```

### Plot 2: Missing Values Overview
Visualizing the amount of missing data across the sensor readings.

```python
plt.figure(figsize=(10, 4))
missing_counts = df.isnull().sum()
plt.plot(missing_counts.values)
plt.title('Missing Values per Feature')
plt.xlabel('Feature Index')
plt.ylabel('Number of Missing Values')
plt.show()
```

### Plot 3: Correlation Heatmap
Checking for high correlations among the first 20 features.

```python
plt.figure(figsize=(10, 8))
# Sample the first 20 numeric columns for readability
corr_matrix = df.select_dtypes(include=[np.number]).iloc[:, :20].corr()
sns.heatmap(corr_matrix, cmap='viridis', annot=False)
plt.title('Feature Correlation Heatmap (First 20 Features)')
plt.show()
```
